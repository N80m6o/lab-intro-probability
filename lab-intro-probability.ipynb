**Challenge 1**

#### Ironhack Airlines 

Often Airlines sell more tickets than they have seats available, this is called overbooking. Consider the following:
- A plane has 450 seats. 
- Based on historical data we conclude that each individual passenger has a 3% chance of missing it's flight. 

If the Ironhack Airlines routinely sells 460 tickets, what is the chance that they have a seats for all passenger?

```python
# Given values
n = 460  # number of tickets sold
p = 0.97  # probability that a passenger shows up
k = 450  # number of seats available

# Calculate the probability that 450 or fewer passengers show up
probability_seats_available = binom.cdf(k, n, p)

print(f"The probability that there are enough seats for all passengers is {probability_seats_available:.4f}")
```

**Challenge 4**

#### Ironhack Helpdesk

Consider a scenario related to the time between arrivals of customers at a service desk.

On average, a customers arrives every 10minutes.

What is the probability that the next customer will arrive within the next 5 minutes?

```python
# Given values
average_time_between_arrivals = 10  # in minutes
lambda_rate = 1 / average_time_between_arrivals  # rate parameter, λ = 1 / average time between arrivals
time = 5  # time for which we want to calculate the probability, in minutes

# Calculate the probability
probability = 1 - np.exp(-lambda_rate * time)

print(f"The probability that the next customer will arrive within the next {time} minutes is {probability:.4f}")
```

If there is no customer for 15minutes, employees can that a 5minutes break.

What is the probability an employee taking a break?

```python
# Given values
average_time_between_arrivals = 10  # in minutes
lambda_rate = 1 / average_time_between_arrivals  # rate parameter, λ = 1 / average time between arrivals
time_no_arrival = 15  # time for which we want to calculate the probability, in minutes

# Calculate the probability that no customer arrives within 15 minutes
probability_no_customer = np.exp(-lambda_rate * time_no_arrival)

print(f"The probability that no customer arrives within {time_no_arrival} minutes, allowing an employee to take a break, is {probability_no_customer:.4f}")
```

**Challenge 5**

The weights of a certain species of birds follow a normal distribution with a mean weight of 150 grams and a standard deviation of 10 grams. 

- If we randomly select a bird, what is the probability that its weight is between 140 and 160 grams?

```python
#code here
# Given values
mean = 150
std_dev = 10

# Z-scores for 140 and 160
z1 = (140 - mean) / std_dev
z2 = (160 - mean) / std_dev

# Probability that the bird's weight is between 140 and 160 grams
probability = norm.cdf(z2) - norm.cdf(z1)

print(f"The probability that the bird's weight is between 140 and 160 grams is {probability:.4f}")
```

**Challenge 6**

If the lifetime (in hours) of a certain electronic component follows an exponential distribution with a mean lifetime of 50 hours, what is the probability that the component fails within the first 30 hours?

```python
#code here
# Given values
mean_lifetime = 50  # in hours
lambda_rate = 1 / mean_lifetime  # rate parameter, λ = 1 / mean lifetime
time = 30  # time for which we want to calculate the probability, in hours

# Calculate the probability that the component fails within the first 30 hours
probability_failure = 1 - np.exp(-lambda_rate * time)

print(f"The probability that the component fails within the first {time} hours is {probability_failure:.4f}")
```
